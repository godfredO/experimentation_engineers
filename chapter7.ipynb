#### Don't evaluate the model, evaluate the product

```python
plt.figure(figsize=(12,8))
prediction_quality = np.linspace(0,1,10)
dau = 1 - np.exp(-prediction_quality)           # daily active users
dau[-2] = dau[-1]                               # last two values are the same, plateau
plt.plot(prediction_quality, dau, 'o', color=e4e.color_1)
plt.xlabel('Prediction quality')
plt.ylabel('DAU')
plt.xticks([])
plt.yticks([])
plt.show()
```
